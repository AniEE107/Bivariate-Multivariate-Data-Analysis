```python
#Numerical-Numerical:
#1.Scatter plot
sns.scatterplot(x=flights['Price'], y=flights['Airline'], hue=flights['Source'], style=flights['Destination'])
```

```python
#2.Bar_Plot:NUMERICAL-CATGORICAL
sns.barplot(x=flights['Price'], y=flights['Destination'], hue=flights['Source'])
```

```python
#HETAMAP:CATEGORAL-CATEGORICAL:
sns.heatmap(pd.crosstab(flights['Airline'],flights['Source']))
```

```python
#Cluster_Map:Categorical-Categorical 
sns.clustermap(pd.crosstab(flights['Destination'],flights['Source']))
```
